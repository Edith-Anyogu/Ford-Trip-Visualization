## Investigation Overview


> In this investigation, I wanted to look at the time duration of bike trips, and the variables that can influence how long a trip will last. Some of the variables I will be looking at are duration, age, hour trip started,gender,user type, and bike sharing.


## Dataset Overview

> The dataset contains 183412 of bike rides from 2019. There are 16 columns, these columns are variables explaining bike rides and users. Some of the variables were cleaned up.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
ford = pd.read_csv('fordtrip.csv')
```

## Distribution of Bike Rides Duration

## (Visualization 1)

> Bike trip duration has a unimodal distribution. Most duration are on the lower limit, with peak duration being 10 minutes.



```python
# put the histogram on a log scale
bins = 10 ** np.arange(0,3 +0.1,0.1)
ticks =  [ 0.1 , 0.3 , 1 , 3, 10, 30, 100, 300]
labels= ['{}'.format(v) for v in ticks]

plt.hist(data=ford,x='duration_in_minutes',bins = bins)
plt.xscale('log')
plt.xticks(ticks,labels);
```

## Time of day that bike ride started

## (Visualization 2)

> The time of the day that bike ride started is likely to influence the duration of trips, from the visualization rides started the 5th hour, just before the 10th hour of the day have a positive correlation with duration and trips started before the 20th hour is also has a positive correlation with duration.

```python
# visualize duration and minutes  trip started
plt.hist2d(data= ford, x='duration_in_minutes',y='hour_start')
plt.colorbar()
plt.title('Hour versus Duration')
plt.xlabel('Duration')
plt.ylabel('Hour');
```

## Age of Bike Users

## (Visualization 3)
Greater number of bike users are young people with an average age 34-35. The young age groups are morelikely to spend more time during bike rides.



```python
# scatter plot of duration vs. age
plt.scatter(data = ford, x = 'age', y = 'duration_in_minutes', alpha = 1/5)
plt.xlabel('Age')
plt.xlim(18, 70)
plt.ylabel('Trip Duration (min)')
plt.title('Duration versus User Age');
```

## Gender, user type and trip duration

## (Visualization 4)

> Though from the data male gender has the highest values, they have short trip duration shown in the visualization below. The gender labelled as other has the longest trip duration though it has the smallest value in the data.

```python
# visualize gender, user type and trip duration
ax = sb.barplot(data= ford, x= 'member_gender',y = 'duration_in_minutes', hue = 'user_type')
ax.legend(loc = 8, ncol =3, framealpha = 1, title = 'user_type')
plt.title('Gender versus User type and trip duration');
```

## Bike sharing, user type and trip duration

## (Visualization 5)

> Customer user type has the highest value in the data, this user type are not bike sharers, and they spend more time during bike rides.

```python
# visualize gender, user type and trip duration
ax = sb.barplot(data= ford, x= 'user_type',y = 'duration_in_minutes', hue = 'bike_share_for_all_trip')
ax.legend(loc = 8, ncol =3, framealpha = 1, title = 'bike_share_for_all_trip')
plt.title('User type versus bike sharing and trip duration');
```
